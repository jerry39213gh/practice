```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
from sklearn.decomposition import NMF
art = pd.read_pickle("data/articles.pkl")
snowball = SnowballStemmer('english')
import random 
```

```python
def tokenize(doc):
    return [snowball.stem(word) for word in word_tokenize(doc.lower())]

content = art.content
cv = TfidfVectorizer(max_features=5000, stop_words='english', tokenizer=tokenize)
mtx = cv.fit_transform(content)
wdindex= cv.vocabulary_
```

```python
v=mtx.toarray()
model = NMF(n_components=10, init='random', random_state=0, max_iter=100)
model.fit_transform(v)
model.components_
```

```python
inv_wdindex = {v: k for k, v in wdindex.iteritems()}
inv_wdindex
```

```python
lst=[]
for i in range(10):
    print [inv_wdindex[j] for j in [arr.argsort()[-10:][::-1] for arr in model.components_][i]]
```
